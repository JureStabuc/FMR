```python
#import libraries
import requests
from bs4 import BeautifulSoup as BS
import time
import json
#selenium automates Chrome browser as whoscored.com detects scrapers
from selenium import webdriver
#import options to add add block extension
from selenium.webdriver.chrome.options import Options
```

```python
#Ad block extension has to be used on whoscored.com with selenium as the page loads faster without ads
#Instructions https://www.reddit.com/r/learnpython/comments/4zzn69/how_do_i_get_adblockplus_to_work_with_selenium/
path_to_extension = "path to the ad_block extension"
chrome_options = Options()
chrome_options.add_argument('load-extension=' + path_to_extension)
driver = webdriver.Chrome("./chromedriver", chrome_options=chrome_options)
driver.create_options()
```

```python
#range of numbers that represent the page number in the link
for i in range(1, 39):
    driver.get("https://www.premierleague.com/tables?co=1&se=54&mw=1-{}&ha=-1".format(i))
    time.sleep(10)
    #navigate to the page
    page = driver.page_source
    content = BS(page, "html.parser")
    #find the table with standings
    table = content.find("tbody", class_ = "tableBodyContainer")
    #find all the rows
    club_row = table.find_all("tr") 
    #information is stored in every second row
    club_row = club_row[::2]
    
    league_table = []
    
    #loop through every row
    for row in club_row:
        #find the information in the matchday table
        position = row.find("span", class_="value").text
        club_name = row.find("span", class_="long").text
        #rename the clubs that the names match those of whoscored.com
        if club_name == "Tottenham Hotspur":
            club_name = "Tottenham"

        elif club_name == "West Ham United":
            club_name = "West Ham"

        elif club_name == "Hull City":
            club_name = "Hull"

        elif club_name == "Leicester City":
            club_name = "Leicester"

        elif club_name == "Manchester City":
            club_name = "Man City"

        elif club_name == "Manchester United":
            club_name = "Man Utd"

        elif club_name == "West Bromwich Albion":
            club_name = "West Brom"

        elif club_name == "Stoke City":
            club_name = "Stoke"

        elif club_name == "Swansea City":
            club_name = "Swansea"

        else:
            club_name = club_name
        #iformation interested in
        played = row.find_all("td")[3].text
        won = row.find_all("td")[4].text
        drawn = row.find_all("td")[5].text
        lost = row.find_all("td")[6].text
        points = row.find("td", class_ = "points").text
        data = {
                "position":position,
                "club_name":club_name,
                "played":played,
                "won":won,
                "lost":lost,
                "drawn":drawn,
                "points":points

        }
        #append the dictionary to the league_table array
        league_table.append(data)
    #open a json file and write all matchday tables as a json string
    with open('../data/path/matchday_table/matchday{}.json'.format(i), "w") as f:
        json.dump(league_table, f)
        time.sleep(5)
    time.sleep(5)       
```
