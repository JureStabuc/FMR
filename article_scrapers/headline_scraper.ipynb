```python
#import libraries
import requests
from bs4 import BeautifulSoup as BS
import time
```

```python
#open a text file to write headlines into
with open("headlines.txt", "w") as file:
    #range of numbers that represent the page number in the link
    for i in range(12):
        #create the link and navigate to it
        page_url = "https://www.theguardian.com/football/premierleague+tone/matchreports?page={}".format(i)
        match_headline = requests.get(page_url).text
        soup = BS(match_headline, "html.parser")
        #find the headlines
        titles = soup.find_all("a", class_ = "u-faux-block-link__overlay js-headline-text")
        for title in titles:
            print(title)
            #write the headline text to the text file
            file.write(title.text+"\n"+"\n")
file.close()
```
