```python
#import needed libraries
import requests
from bs4 import BeautifulSoup as BS
import time
```

```python
urls=[]
#defined range which will form the needed id in the link of the report
for j in range(450628,451008):
    page_url = "http://www.espn.co.uk/football/report?gameId={}".format(j)
    #appending the link to the array
    urls.append(page_url)

#open a text file where the match reports will be stored
with open("corpus.txt", "w") as file:
    for link in urls:
        print(link)
        #navigate to the page
        match_page = requests.get(link).text
        soup = BS(match_page, "html.parser")
        try:
            #find the article body and paragraphs
            container = soup.find("div", class_="article-body")
            paras = container.find_all("p")
            #loop through found paragraphs and write them to the text file
            for para in paras:
                file.write(para.text+"\n"+"\n")
                time.sleep(1)
            time.sleep(1)
        except:
            continue

#close the file
file.close()
```
